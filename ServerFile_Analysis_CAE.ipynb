````python
from statsmodels.tsa.stattools import adfuller as ADF
````

````python
diff=0
````

````python
adf=ADF(new_data.访问周期.astype('timedelta64[s]'))
````

````python
while adf[1]>=0.05:
    diff=diff+1
    adf=ADF(new_data.访问周期.astype('timedelta64[s]').diff(diff).dropna())
print(diff,adf[1])
````

````python
from statsmodels.stats.diagnostic import acorr_ljungbox
````

````python
[[lb],[p]]=acorr_ljungbox(new_data.访问周期.astype('timedelta64[s]'),lags = 1)
````

````python
if p<0.05:
    print(u'非白噪声对应的p值为: %s' %p)
else:
    print(u'白噪声对应的p值为: %s' %p)
````
