```python
wordcloud=open(r"C:\Users\awangg2\Documents\wordcloud.txt",'w')
result=[]
for i in file:
    m=re.sub(patt,"",i)
    seg_list = jieba.cut("".join(m))
    for l in seg_list :#读取每一行分词  
        #result.append(w) 
        wordcloud.write(l+'\n')
        #print(l)
wordcloud.close()

```
