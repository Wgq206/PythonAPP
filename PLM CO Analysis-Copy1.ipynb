```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import seaborn as sns
```

```python
df=pd.read_excel(r"C:\Users\awangg2\Desktop\20170221CO.xlsm")
```

```python
df[u'创建日期'] = pd.to_datetime(df[u'创建日期'])
df[u'发放日期']=pd.to_datetime(df[u'发放日期'])
```

```python
data=df[(df[u'创建日期']>='20160101')&(df[u'创建日期']<='20170221')]
```

```python
CO_count=data.groupby([u'所有权项目']).count()
```

```python
CO_sort=CO_count.sort_values(u'当前 ID',ascending=False)
```

```python
CO_project_count=CO_sort[u'当前 ID']
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
CO_project_count.plot(kind='bar')
plt.show()
```

```python
CO_sort[u'发放完成率']=(CO_sort[u'发放日期'])/(CO_sort[u'创建日期'])
```

```python
CO_accuracy=CO_sort.sort_values(u'发放完成率',ascending=False)
```

```python
CO_accuracy_data=CO_accuracy[u'发放完成率']
```

```python
CO_accuracy_data.plot(kind='bar')
plt.show()
```
