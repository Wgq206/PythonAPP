````python
def a3(dataframe, threshold=3.5):
  dd = dataframe['访问周期']
  MAD = (dd - dd.median()).abs().median()
  zscore = ((dd - dd.median())* 0.6475 /MAD).abs()
  dataframe['isAnomaly'] = zscore > threshold
  return dataframe
````
