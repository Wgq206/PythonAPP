## 5.提取数据存放文件夹

````python
import re #导入Re模块
````

````python
pattern=re.compile(r'[^\\]+') #建立分离文件路径的样式，只要是被\分开的就是文件夹
pathList=[]
````

````python
for i in df[u'文件路径']:
    pathList.append(pattern.findall(i))
new1=pd.DataFrame(pathList)
````

````python
#pathList=[pattern.findall(i) for i in df[u'文件路径']] #获取文件夹列表
````

````python
#new=[pd.Series(i)for i in pathList] #转化为Series
````

````python
#new1=pd.DataFrame(new) #转化为Dataframe
````

````python
new_data=pd.merge(df,new1,left_index=True,right_index=True,how='outer') 
#将文件夹列表添加到原始数据中
````
